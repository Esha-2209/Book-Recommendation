```python
ratings_with_books = ratings.merge(df ,on = "ISBN")
```

```python
ratings_with_books.head(2)
```

```python
ratings_with_books.shape
```

```python
ratings_with_books.rename(columns = {
    "Book-Rating":"rating"
},inplace =True)
```

```python
ratings_with_books.head(2)
```

```python
num_rating = ratings_with_books.groupby('Book-Title')['rating'].count().reset_index()
```

```python
num_rating.head()
```

```python
num_rating.rename(columns = {
    "rating":"no of ratings"
},inplace = True)
```

```python
num_rating.head()
```

```python
final_rating = ratings_with_books.merge(num_rating, on ='Book-Title')
```

```python
final_rating.head(2)
```

```python
final_rating.shape
```

```python
final_rating = final_rating[final_rating['no of ratings']>=50]
```

```python
final_rating.head(2)
```

```python
final_rating.sample(10)
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['User-ID','Book-Title'], inplace=True)
```

```python
final_rating.shape
```

# Converting to pivot table

```python
df_pivot = final_rating.pivot_table(columns='User-ID',index = 'Book-Title', values = 'rating')
```

```python
df_pivot.head(5)
```

```python
df_pivot.shape
```

```python
df_pivot.fillna(0, inplace = True)
```

```python
df_pivot
```

# Model building

```python
from scipy.sparse import csr_matrix
```

```python
df_sparse = csr_matrix(df_pivot)
```

```python
df_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm = 'brute')
```

```python
model.fit(df_sparse)
```

```python
distance,suggestion = model.kneighbors(df_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distance
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
    print(df_pivot.index[suggestion[i]])
```

```python
import pickle
pickle.dump(model, open('artifacts/model.pkl','wb'))
pickle.dump(df_name, open('artifacts/df_name.pkl','wb'))
pickle.dump(final_rating, open('artifacts/final_rating.pkl','wb'))
pickle.dump(df_pivot, open('artifacts/df_pivot.pkl','wb'))
```

# Recommendation Process

```python
def recommend_book(df_name):
    book_id = np.where(df_pivot.index == df_name)[0][0]
    distance,suggestion = model.kneighbors(df_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=6)

    for i in range(len(suggestion)):
        books = df_pivot.index[suggestion[i]]
        for j in books:
            print(j)
```

```python
df_name = 'Wish You Well'
recommend_book(df_name)
```
